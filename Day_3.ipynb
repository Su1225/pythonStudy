```python
"""
使用input函数输入
使用int()进行类型转换
用占位符格式化输出的字符串
"""

a = int(input('a ='))
b = int(input('b ='))

#First Style

print('%d + %d = %d' %(a, b, a+b))
print('%d - %d = %d' %(a, b, a-b))
print('%d * %d = %d' %(a, b, a*b))
print('%d / %d = %d' %(a, b, a/b))
print('%d // %d = %d' %(a, b, a//b))
print('%d %% %d = %d' %(a, b, a%b))
print('%d ** %d = %d' %(a, b, a**b))

#Second Style

print('{} + {} = {}'.format(a, b, a+b))
print('{} - {} = {}'.format(a, b, a-b))
print('{} * {} = {}'.format(a, b, a*b))
print('{} / {} = {}'.format(a, b, a/b))
print('{} // {} = {}'.format(a, b, a//b))
print('{} % {} = {}'.format(a, b, a%b))
print('{} ** {} = {}'.format(a, b, a**b))
```

```python
"""

将华氏温度转换为摄氏温度
F = 1.8C + 32
C = (32 - F)/1.8

"""

f = float(input('Give me a temp: '))
c = (32 - f) / 1.8

print('%.1f华氏度 = %.1f摄氏度' % (f, c))
print('{}华氏度 = {}摄氏度'.format(f, c))
```

```python
"""

输入半径计算圆的周长和面积

C = 2piR
S = piR^2
"""

import math

radius = float(input('Please enter the radius: '))
c = 2 * math.pi * radius
s = math.pi * radius * radius

print('Perimeter of this circle is %.2f' % c)
print('Area of this circle is %.2f' % s)


print('Perimeter of this circle is {}'.format(c))
print('Area of this circle is {}'.format(s))

print('Perimeter of this circle is {:.2f}'.format(c))
print('Area of this circle is {:.2f}'.format(s))
```

```python
"""

输入年份 如果是闰年输出True 否则输出False

"""

year = int(input('Please enter a year: '))
is_leap = (year % 4 == 0 and year % 100 != 0 or
           year % 400 == 0 )
if is_leap:
    print('{} is  a leap year'.format(year))
else:
    print('{} is  not a leap year'.format(year))
```

```python
"""

分段函数求值

        3x - 5  (x > 1)
f(x) =  x + 2   (-1 <= x <= 1)
        5x + 3  (x < -1)

"""


x = float(input('enter the x = '))
if x > 1:
    y = 3*x - 5
elif -1 <= x and x <= 1:
    y = x +2
else:
    y = 5 * x + 3
    
print('f(%.2f) = %.2f' % (x, y))    
print('f({:.2f}) = {:.2f}'.format(x,y))


x = float(input('x = '))
if x > 1:
    y = 3 * x - 5
elif x >= -1:
    y = x + 2
else:
    y = 5 * x + 3
print('f(%.2f) = %.2f' % (x, y))
```

```python
"""

英制单位英寸和公制单位厘米互换

"""

length = float(input('Enter the length: '))
unit = input('Enter the unit: ')

if unit == 'inch' or unit == 'in':
    print('{} inch = {} centimeter'.format(length, length * 2.54))
elif unit == 'centimeter' or unit == 'cm' or unit == 'cent':
    print('{} cent = {} inch'.format(length, length / 2.54))
else:
    print('Please enter the vaild unit')
```

```python
"""

判断输入的边长能否构成三角形
如果能则计算出三角形的周长和面积

"""
import math

a = int(input('first side len: '))
b = int(input('second side len: '))
c = int(input('third side len: '))

if a + b > c and a + c > b and b + c > a:
    perimeter = a + b + c
    p = perimeter / 2
    area = math.sqrt(p * (p - a)*(p - b)*(p - c))
    print('This triangle\'s perimeter is {}, and it\'s area is {}'.format(perimeter, area))
else:
    print('This is not a triangle')
```

```python
"""

输入月收入和五险一金计算个人所得税

"""

salary = float(input('Enter your monthly salary: '))
social_insurance = salary * 0.104 
house_insurance = salary * 0.12

insurance = social_insurance + house_insurance
diff = salary - insurance - 5000

if diff <= 0:
    rate = 0
    deduction = 0
elif diff < 3000:
    rate = 0.03
    deduction = 0
elif diff < 12000:
    rate = 0.1
    deduction = 210
elif diff < 25000:
    rate = 0.2
    deduction = 1410
elif diff < 35000:
    rate = 0.25
    deduction = 2660
elif diff < 55000:
    rate = 0.3
    deduction = 4410
elif diff < 80000:
    rate = 0.35
    deduction = 7160
else:
    rate = 0.45
    deduction = 15160
tax = abs(diff * rate - deduction)
print('your insurance: ￥{:.2f} RMB,social insurance: ￥{:.2f} RMB and house insurance: ￥{:.2f} RMB'.format(insurance, social_insurance, house_insurance))
print('personal taxes: ￥{:.2f} RMB'.format(tax))
print('real salary: ￥{:.2f}'.format(diff + 5000 - tax))
```
